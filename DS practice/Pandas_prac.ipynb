```python
# Binary  operation on series:
"""
We can perform binary operation on series like addition, subtraction and many other operation. 
In order to perform binary operation on series we have to use some function like .add(),.sub() etc..

"""
import numpy as np 
import pandas as pd
dt = pd.Series([1,2,3,4],index=['a','d','r','g']) 
dt1 = pd.Series([9,8,7,6],index=['z','x','c','v'])

data = dt.add(dt1,fill_value=0)  # Adding two series
print(data)

data1 = dt.sub(dt1,fill_value=0)   # Sub two series 
print(data1)

#The fill_value parameter in dt.sub(dt1, fill_value=0) specifies the value that should be 
# used to fill any missing values resulting from the alignment of the two DataFrames or Series
```

```python
# Conversion in series:
"""
In conversion operation we perform various operation like changing datatype of series, changing
 a series to list etc. In order to perform conversion operation we have various function which help 
 in conversion like .astype(), .tolist() etc.
"""
import pandas as pd 
data = pd.read_csv("Sample100.csv")
print(data)
before = data.dtypes 
print(before)

data["Number"]=data["Number"].astype(int)
data["Employee Name"]=data["Employee Name"].astype(str)
after = data.dtypes
print("Before conversion \n",before,"\n")
print("After conversion\n",after,"\n")

before_dtype_salary = type(data["Number"])
print(before_dtype_salary)
after_dtype_salary = data["Number"].tolist()

print("Before\n",before_dtype_salary,"\n")
print("After\n",after_dtype_salary,"\n")
```
