```python
import pandas as pd 
import numpy as np 
import scipy.stats as stats

car=pd.read_csv('mtcars.csv')
```

```python
car.describe()
```

```python
#Calculating the range of the mpg and disp 
print('Range of mile per gallon:', np.ptp(car['mpg']))
print('Range of displacement:', np.ptp(car['disp']))
```

```python
# Calculating the IQR using the np 
stats.iqr(car[['mpg', 'disp', 'wt']],axis=0, interpolation='lower')
```

# Measure of dispersion 

1. Range 
2. variance 
3. Standard deviation
4. **IQR**
5. **Skewness 
6. Kurtosis **

The last three should be calculate using stats, other can be calculated using np 


```python
np.std(car[['mpg','disp','wt']])
```

```python
np.var(car[['mpg','disp','wt']])
```

```python
from scipy.stats import skew
skew(car[['mpg','disp','wt']])
```

```python
from scipy.stats import kurtosis
kurtosis(car[['mpg','disp','wt']])
```

```python
# to look at it as graph wise as below 
import matplotlib.pyplot as plt 
import seaborn as sns
plt.figure(figsize=(30,10)) 
with plt.style.context('dark_background'):
    plt.subplot(1,3,1)
    sns.histplot(car['mpg'],kde=True,color='Red')
    plt.subplot(1,3,2)
    sns.histplot(car['disp'],kde=True,color='cyan')
    plt.subplot(1,3,3)
    sns.histplot(car['wt'],kde=True,color='green')
```

# Calculating meassures oif position 


```python
import pandas as pd 
import numpy as np 
import scipy.stats as stats
from scipy.stats import zscore 

car=pd.read_csv('mtcars.csv')
```

```python
# Calculating the 50 percentile of the mpg 

np.percentile(car['mpg'],50, axis=0)
```

```python
#Calculating quantiles 
numeric_column=car.select_dtypes(include=['number'])

numeric_column.quantile([0.25,0.5,0.75])
```

```python
#Calculating Zscore 

zscore(car.mpg,axis=0)
```

# Super market study:



```python
import numpy as np 
import pandas as pd 
from scipy.stats import iqr,skew
```

```python
data=pd.read_csv('supermarket.csv')
```

```python
data
```

```python
data.describe()
```

```python
#Calculating mean and median for the amount 

print('Mean of the Amount:', round(data.Amount.mean(),2))
print('--------------------------')
print('Median of the Amount:', round(data.Amount.median(),2))
```

```python
# Calculating the measures of despersion:
print('Range in which the amount is varying:', np.ptp(data.Amount))
print('-------------------------------------------------')
print('Standard deviation of the amount:', np.std(data.Amount))
print('-------------------------------------------------')
print('The inter quartile range is:',iqr(data.Amount))
print('-------------------------------------------------')
print('The variance is:', np.var(data.Amount))
```

```python
#Calculating the measures of the position 

print('The 20th percentile of the amount is:', np.percentile(data.Amount,20))
print('-------------------------------------------------')
print('The quantile is:', np.quantile(data.Amount,0.75))
print('-------------------------------------------------')
from scipy.stats import zscore 
print('The standard score is:',zscore(data.Amount,axis=0))
```
