# Descriptive Statistics
<h2> Demo 3: Calculate Measures of Position using Python </h2>

In this demo, you will be shown how to calculate the measures of postion such as percentiles, quartiles and z-scores.

```python
#Step1: Import the required libraries
import pandas as pd
import numpy as np         #for calculating percentile
import scipy.stats         #for calculating z-score
```

```python
#Step2: Read data from the uploaded csv file
cars = pd.read_csv("mtcars.csv")
```

```python
#Step3: Display the dataframe
cars
```

```python
#Step4: Calculate the percentile
np.percentile(cars[['mpg','disp']], 50, axis=0)       #50th percentile of mpg and disp columns
```

```python
#Step5: Calculate the quartile
cars.quantile(0.25)                                #1st quartile or quantile of order 0.25
```

```python
#Step6: Calculate the z-score
from scipy.stats import zscore
zscore(cars.mpg, axis=0)       #z-score of each element in the mpg column
```
