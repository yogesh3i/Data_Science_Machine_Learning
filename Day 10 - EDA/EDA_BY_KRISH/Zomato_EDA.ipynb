```python
# There are simillar columns in the datasets thats country codes we can merge both the dataset 

final_df=pd.merge(df,df_contry,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
# check data types

final_df.dtypes
```

```python
# checking total cities served in perticular country 

final_df.Country.value_counts()
```

```python
# Its making an series of different records in the columns and having an index of the country name 
# hence when we are giving the index at end its providing us the index names which is country 

country_name=final_df.Country.value_counts().index
```

```python
# now let just get the values of the records only. How many values are there in the each index i.e. country 
Country_val=final_df.Country.value_counts().values
```

```python
 # now let just plot this two variables in the pie chart for top3 only 

plt.pie(Country_val[:3],labels=country_name[:3],autopct='%1.2f%%')
plt.show()
```

Observation from the graph: 
1. Most of the zomato users are there in the india,followed by USA and UK.



```python
final_df.columns
```

```python
# Now lets look at the numerical values in the data, ratings 

final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()   # ot will give me dataframe 
```

```python
# If i do the reset index over thr above query it will give me an dataframe 
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()
```

```python
# Now lets rename the abive 0 names column
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating_count'})
```

# Observation: 
1. When rating is between 4.5 to 4.9 the rating is excelent 
2. When rating are between 4.0 to 4.4  its very good rating 
3. When rating is between 3.5 to 3.9 its  good rating 
4. From 2.5 to 3.4 its average rating 
5. From 1.8 to 2.4 its poor 

```python
ratings.head()
```

```python

matplotlib.rcParams['figure.figsize']=(12,8)
# fig=plt.figure(figsize=(12,8))
sns.barplot(x='Aggregate rating', y='Rating_count',hue='Rating color',data=ratings,width=2,capsize=0.4)
plt.tight_layout()
plt.legend(loc='upper right')
plt.show()
```

```python
# but in the above figure the colors are not matvhing corectly to do that we can use palette 
with plt.style.context('dark_background'):
    matplotlib.rcParams['figure.figsize']=(12,8) 
    # fig=plt.figure(figsize=(12,8))
    sns.barplot(x='Aggregate rating', y='Rating_count',hue='Rating color',data=ratings,width=2,capsize=0.4,palette=['white','red','orange','yellow','green','green'])
    plt.tight_layout()
    plt.legend(loc='upper right')
    plt.show()
```

Observation: 
1. Not rated count is very high.
2. Maximum number of rating is between 2.5 to 3.4 

```python
# we can plot the count plot of the rating i.e. how much red and other ratings are given 

sns.countplot(x='Rating color',data=ratings,palette=['white','red','orange','yellow','green','green'])
plt.show()
```
