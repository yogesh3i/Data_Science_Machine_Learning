# EDA using pie chart: Dmeo-1 

```python
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np 
```

```python
# Performing for fruits stocks 

fruits=['Orange','Apple','Guava','watermelon','Papaya']
stocks=[33,20,19,9,25]

labels=fruits
colors=['#FF8C00','#8B0000','#FF1493','#ADFF2F','#9400D3']
```

```python
# plotting pie chart 
plt.pie(stocks,labels=labels,colors=colors)
plt.show()
```

```python
# plotting percentage:
plt.pie(stocks,labels=labels,colors=colors,autopct='%0.1f%%',textprops={'weight':'bold','color':'black'})
plt.title('Available stock of fruits in MY Farm')
plt.show()
```

```python
# Exploding the maximum available fruits quantity 
explode=(0.1,0,0,0,0)
plt.pie(stocks,labels=labels,colors=colors,autopct='%0.1f%%',explode=explode,textprops={'weight':'bold','color':'black'})
plt.title('Available stock of fruits in MY Farm')
plt.show()
```

```python
# adding shadow to the graph 

plt.pie(stocks,labels=labels,colors=colors,autopct='%0.1f%%',explode=explode,shadow=True,textprops={'weight':'bold','color':'CYAN'})
plt.title('Available stock of fruits in MY Farm')
plt.show()
```

```python
# Selecting the start angle, the ususal start angle is 90 

plt.pie(stocks,labels=labels,colors=colors,autopct='%0.1f%%',explode=explode,startangle=180,shadow=True,textprops={'weight':'bold','color':'black'})
plt.title('Available stock of fruits in MY Farm')
plt.show()
```

```python
# adding frame parameter to the graph which will show the axis around it 

plt.pie(stocks,labels=labels,colors=colors,autopct='%0.1f%%',explode=explode,startangle=180,shadow=True,frame=True,textprops={'weight':'bold','color':'black'})
plt.title('Available stock of fruits in MY Farm',)
plt.show()
```

```python
# Also i can rotate the labels of the chart along to the quantity and the name 

plt.pie(stocks,labels=labels,colors=colors,autopct='%0.1f%%',explode=explode,startangle=180,shadow=True,rotatelabels=True,textprops={'weight':'bold','color':'black'})
plt.title('Available stock of fruits in MY Farm')
plt.show()
```

```python
# With the help of the wedgeprops i can change the tickness of the edges in the chart 

plt.pie(stocks,labels=labels,colors=colors,autopct='%0.1f%%',explode=explode,startangle=180,shadow=True,rotatelabels=True,textprops={'weight':'bold','color':'green'},
wedgeprops={'linewidth':4})

plt.show()
```

```python
# I can also decide the size of the chart by giving radius

plt.pie(stocks,labels=labels,colors=colors,autopct='%0.1f%%',explode=explode,startangle=180,shadow=True,rotatelabels=True,radius=3,textprops={'weight':'bold','color':'#C71585'},
wedgeprops={'linewidth':4})

plt.show()
```

```python
# Changing the orientation of the graph 
plt.hist(values,histtype='bar',color='c',edgecolor='w',orientation='horizontal',label='Test score data')
plt.title('Histogram of the score',fontweight='bold')
plt.xlabel('Score',fontweight='bold')
plt.ylabel('Frequency',fontweight='bold')
plt.legend()
plt.show()
```

```python
# Plotting two differtent distribution along side and comparing both 
values1 = [42,36,24,40,67,62,45,58,41,32,48,49,38,57,42,72,57,56,56,43]
plt.hist(values,bins=10, edgecolor='black',label='Test score data')
plt.hist(values1,bins=10,edgecolor='black',label='Test1 score data')
plt.title('Histogram of score')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Change the transparencies of the overlapping bars by using alpha
values1 = [42,36,24,40,67,62,45,58,41,32,48,49,38,57,42,72,57,56,56,43]
plt.hist(values,bins=10, edgecolor='black',label='Test score data',alpha=1)
plt.hist(values1,bins=10,edgecolor='black',label='Test1 score data',alpha=0.5)
plt.title('Histogram of score')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
# Also can change the color 

values1 = [42,36,24,40,67,62,45,58,41,32,48,49,38,57,42,72,57,56,56,43]
plt.hist(values,bins=10, edgecolor='black',label='Test score data',alpha=1,color='c')
plt.hist(values1,bins=10,edgecolor='black',label='Test1 score data',alpha=0.5,color='r')
plt.title('Histogram of score')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

# Bivariate EDA- Here we plot two or more variable:Demo 5

# Using Non-graphical technique first.

Plotting relationship between the two or more variables allows us to get the visual understanding of the patterns and the corelation easily.

Bivariate Exploratory Data Analysis (EDA) is a branch of data analysis that focuses on examining the relationship between two variables in a dataset. In bivariate EDA, we investigate how two variables in the data are related to each other and how changes in one variable might influence the other.

**Its like univariate analysis only it involves both the computing summary statistics as well as producing the visual displays.** 

**The appropriate type of the bivarient depends on the nature of the data: numerical vs. categorical**


# 1.Numerical vs. Numerical (Continuous or Discrete):

If both variables are numerical, you would typically use techniques like scatter plots and correlation analysis to explore the relationship between them. Scatter plots are useful for visualizing the distribution and patterns of two numerical variables and identifying potential correlations or trends. Correlation analysis, such as calculating Pearson's correlation coefficient or Spearman's rank correlation, quantifies the strength and direction of the linear relationship between two numerical variables.

# 2.Categorical vs. Categorical (Nominal or Ordinal):
When dealing with two categorical variables, you can use cross-tabulation (contingency tables) to display the frequency distribution of the combinations of categories. This helps to identify any associations or dependencies between the two categorical variables. You might also use visualization techniques like stacked bar charts or heatmaps to display the joint distribution and identify patterns in the data.

# 3.Numerical vs. Categorical:
If you have one numerical and one categorical variable, you can use techniques like box plots or violin plots to visualize the distribution of the numerical variable across different categories of the categorical variable. These plots provide insights into how the numerical variable varies across different groups and help identify potential differences or relationships between the groups.

```python
# Else i can find out the outliers using zscore, because if the zscore is falling beyond 3rd standard deviation its an outlier definately 

outliers=[]
def detect_outliers(data):
    threshold=3
    mean=np.mean(data)
    std=np.std(data)

    for i in data:
        z_score=(i-mean)/std
        if z_score>threshold:
            outliers.append(i)

    return outliers
```

```python
import seaborn as sns 
dat=sns.load_dataset('iris')
```

```python
detect_outliers(dat['petal_length'])
```
