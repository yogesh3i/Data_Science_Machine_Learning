```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
# Reading data with pandas 
fifa = pd.read_csv('players_20.csv')
```

```python
fifa.head()
```

```python
for cols in fifa.columns:
    print(cols)
```

```python
# check shape of the data 

fifa.shape
```

```python
# checking the nationality of players,get to know from which country how many players are there 

fifa['nationality'].value_counts()   # if you eanted to know more about catagorical data use value_counts()
```

```python
fifa['nationality'].value_counts()[0:10]
```

```python
fifa['nationality'].value_counts()[0:5]
```

```python
fifa['nationality'].value_counts()[0:5].keys()
```

```python
# plotting bar graph for nationality 
fig=plt.figure(figsize=(8,5))
fig.set_facecolor('white')
plt.bar(list(fifa['nationality'].value_counts()[0:5].keys()),list(fifa['nationality'].value_counts()[0:5]),color='c')
plt.xlabel('Country', color='black',fontsize=10,fontweight='bold')
plt.ylabel('Number_of_players', color='black',fontsize=10,fontweight='bold')
plt.title('Players_from_each_country',color='black',fontsize=10,fontweight='bold')
```

```python
# Extracting salary of players 

players_salary=fifa[['short_name','wage_eur']]
```

```python
players_salary.head()
```

```python
# sorting the players accorfing to their salarys 

players_salary=players_salary.sort_values(by=['wage_eur'],ascending=False)
```

```python
players_salary.head() # Now its in ascending order 
```

```python
# Plotting bar graph for it 
fig=plt.figure(figsize=(8,6))
fig.set_facecolor('orange')
plt.bar(list(players_salary['short_name'])[0:5],list(players_salary['wage_eur'])[0:5],color='m')
plt.xticks(rotation=90)
plt.xlabel('Players_name',color='r')
plt.ylabel('Salary',color='r')
plt.show()
```
