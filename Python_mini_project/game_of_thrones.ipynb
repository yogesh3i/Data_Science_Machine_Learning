```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
battle=pd.read_csv('battles.csv')
```

```python
battle.head()
```

```python
# checking shape of the data 
battle.shape
```

```python
# checking null values in the data
battle.isnull()
```

```python
# filling null values in the data 
battle=battle.fillna(0)
```

```python
battle.head()
```

```python
# Renaming attacker_1 as primary attacker 
battle.rename(columns={'attacker_1':'primary_attacker'},inplace=True)
```

```python
# Renaming defender_1 as primary_defender

battle.rename(columns={'defender_1':'primary_defender'},inplace=True)
```

```python
battle.head()
```

```python
# Who attackes most
battle['attacker_king'].value_counts()
```

```python
# At what place the war happened most of the times 

battle['location'].value_counts()
```

```python
# Visualizing the the attacker and the attacker size 

sns.set(rc={'figure.figsize':(13,5)})
sns.barplot(x='attacker_king',y='attacker_size',data=battle)
plt.xticks(rotation=90)
plt.show()
```

```python
# visualizing the defender and the defender size 

sns.set(rc={'figure.figsize':(15,12)})
sns.barplot(x='defender_king',y='defender_size',data=battle)
plt.xticks(rotation=90)
plt.show()
```

```python
# lets see what kind of baattle the king are engaged in 
sns.countplot(x=battle['attacker_king'],hue=battle['battle_type'])
```
