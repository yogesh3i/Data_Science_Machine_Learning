```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
# Reading data
marvel=pd.read_csv('marvel.csv')
```

```python
marvel.head()
```

```python
marvel.shape
```

```python
marvel['Alignment_x'].value_counts()
```

```python
# looking at the good characters 

good= marvel[marvel['Alignment_x']=='good']
```

```python
good.head()
```

```python
# gettigng speed in the ascending order 

good.sort_values(by=['Speed'],ascending=False).head()
```

```python
good_max_speed=good[good['Speed']==10]
```

```python
good_max_speed.head()
```

```python
# Min. speed of the char. 

min_speed=good[good['Speed']==1]
```

```python
min_speed.head()
```

```python
# Sorting by power


good.sort_values(by=['Power'],ascending=False).head()
```

```python
# Sorting with good max power
good_max_power=good.sort_values(by=['Total'],ascending=False)
```

```python
good_max_power.head()
```

```python
# plotting bar graph for total max power 
plt.bar(list(good_max_power['Name'])[0:8],list(good_max_power['Total'])[0:8],color=['m','r','b','y','g','maroon','black','c'])
plt.xticks(rotation=90)
plt.show()
```

```python
# extracting the bad alignment:

bad =marvel[marvel['Alignment_x']=='bad']
```

```python
bad.head()
```

```python
# checking the bad speed 
bad.sort_values(by=['Speed'],ascending=True).head()
```

```python
# checking bad intelligence 
bad.sort_values(by=['Intelligence_y'],ascending=False).head()
```

```python
# checking total 
bad.sort_values(by=['Total'],ascending=False).head()
```

```python
# extracting bad_max_power
bad_max_power=bad.sort_values(by=['Total'],ascending=False)
```

```python
bad_max_power.head()
```

```python
plt.bar(list(bad_max_power['Name'])[0:5],list(bad_max_power['Total'])[0:5],color=['m','r','y','g','c'])
```

```python
# checking speed of the good char. with the help of histogram 

plt.figure(figsize=(8,8))
plt.hist(good['Speed'],color='m')
plt.title('Distribution of the speed')
plt.xlabel('Speed')
plt.show()
```
