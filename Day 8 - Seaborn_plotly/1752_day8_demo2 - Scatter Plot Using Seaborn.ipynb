#Data Visualizaion Using Seaborn

###Demo 2 - Scatter Plot Using Seaborn

####A Data Analysis is to be perfomed on the Iris Dataset, which is the available dataset of Seaborn Library. The Iris Dataset considered here consists of three species- Setosa, Virginica and Versicolor. The Species has 4 important features like sepal_length, sepal_width, petal_width and petal_width. The relationship between the features needs to visualized properly in order to understand the features in depth.

```python
#import the necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Since iris dataset is available in the Seaborn Library, it is loaded using the load_dataset()
df = sns.load_dataset('iris')
```

```python
#reading the first five rows of the iris dataset
df.head()
```

```python
# checking for null values in the dataframe
df.isnull().sum()
```

## Creation of Scatter Plot

```python
#since the dataset contains no null values, therefore Scatter Plot can be used for visualization
sns.scatterplot(data = df, x= 'sepal_length', y='petal_length', hue = 'species')
```

#####The above plot shows that the species setosa of iris dataset has a sepal length which varies from 4 to 6 and the petal length which varies from 1 to 2.Similarly the species versicolor has sepal length varying from 5 to 7 and petal length from 3 to 5. Moreover, the Virginica species has sepal length between 5.7 to 8 petal length between 4.5 to 7.

```python
sns.scatterplot(x = "sepal_width", y='petal_width', data = df, style='species', hue = 'species')
```

####A similar plot is created with the above three species using the style parameter  of scatter plot. The benefit of using hue is that it gives a proper distinction among the various species thus allowing us to distinguish and analysis all the three species

```python
sns.scatterplot(x='sepal_length', y='sepal_width', data=df, hue='species', palette='pastel')
```

####using the palette the points can be created with different colors. In the above example the color pastel has been used.

```python
sns.scatterplot( x = 'petal_length', y='species', data = df, legend='brief')
```

####using the parameter legend the data is represented with evenly spaced values. In this case the value brief is used for the legend.

```python
sns.scatterplot(x='sepal_length', y='species', data=df, alpha = 0.3)
```

####The proportional opacity of the points can be controlled with the help of alpha parameter.Here the alpha value 0.3 is used

#### The above visualization is done in order to analyze the relationship between the features of the three species of Iris flower. Scatter Plot is used in this case as it can properly show the relationship between any two features present in the dataset.
