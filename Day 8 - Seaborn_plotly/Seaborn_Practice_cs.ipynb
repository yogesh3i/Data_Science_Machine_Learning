```python
f,ax= plt.subplots(figsize=(20,10))
sns.lineplot(x='JobRole',y='MonthlyIncome',data=df,ax=ax)

"""plt.subplots() is a function that returns a tuple containing a figure and axes object(s). Thus when using 
fig, ax = plt.subplots() you unpack this tuple into the variables fig and ax. Having fig is useful if you want 
to change figure-level attributes or save the figure as an image file later (e.g. with fig.savefig('yourfilename.png')). 
You certainly don't have to use the returned figure object but many people do use it later so it's common to see. 
Also, all axes objects (the objects that have plotting methods), have a parent figure object anyway, thus:

fig, ax = plt.subplots()
is more concise than this:

fig = plt.figure()
ax = fig.add_subplot(111) """
```

```python
f,ax = plt.subplots()
new_df = df['Gender'].groupby(df['Attrition']).value_counts(normalize=True).rename('Percentage of group').reset_index()

sns.barplot(x='Attrition',y='Percentage of group',hue='Gender',data=new_df)
ax.set(title='Distribution of Gender')
vals= ax.get_yticks()
ax.set_yticklabels(['{:,.0%}'.format(x) for x in vals]) 

"""What does the format string "{:.0%}" mean?

:% â€“ The colon symbol followed by the percentage symbol means â€œconvert it to a percentage numberâ€.
.0 â€“ The dot symbol followed by the zero means â€œuse zero digits after the decimal numberâ€. """
```

```python
#### **10. What is Percentage of Research Scientist in IBM?* 
# Its catagorical data so we will use pie chart to see this. 

labels= df.JobRole.value_counts().index
colors=["olive","orange","hotpink","slateblue","y","Lime"]
sizes= df.JobRole.value_counts().values
plt.figure(figsize=(8,8))
plt.pie(sizes,labels=labels,colors=colors,autopct="%1.1f%%")
plt.title('Percentage of Research Scientist',color='Pink',fontsize=20)
```
