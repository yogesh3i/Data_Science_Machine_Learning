#Data Visualizaion Using Seaborn

###Demo 1 - Heat Map Using Seaborn

#### An analysis on the flights dataset is performed using the Heat Map in order to check the number of passengers travelling during a particular time in a particular year. Different variation of Heat Map is used in order to analyze the total number of passengers travelling on flights.

```python
#importing necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
##importing flights dataset, which is one of the available datasets in Seaborn Library
df = sns.load_dataset('flights')
```

```python
#reading the first five rows of the dataset
df.head()
```

```python
#checking for null values in the dataset
df.isnull().sum()
```

#### Since the no null values are found, analysis through Heat Map can be performed now

## Creation of Heat Map

```python
df = df.pivot("month", "year", "passengers")
sns.heatmap(df)
```

```python
## cmap is used to create colormaps, the cmap used here is tab20
cmap = "tab20"
  
# plotting the heatmap
sns.heatmap(data=df,
                cmap=cmap)
  
```

```python
##annotating each cell with numeric values
sns.heatmap(df, annot=True, fmt="d")
```

```python
#the thickness and color of the lines can be changed with the help of linewidths and linecolor parameters 
linewidths = 2
linecolor = "pink"
sns.heatmap(data=df,
                linewidths=linewidths,
                linecolor=linecolor)
```

####In the above analysis it is found with the help of Heat Map that the number of passengers that were travelling during the month of July and August in 1960 was the highest and the number of passengers travelling during the year 1949 was the lowest. This can be seen in all the graphs that are visualized above.
