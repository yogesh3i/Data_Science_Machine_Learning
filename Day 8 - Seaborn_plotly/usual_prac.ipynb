```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import warnings 
warnings.filterwarnings('ignore')

df=pd.read_csv('IBM.csv')
df.head()
```

```python
df.dtypes
```

```python
figure=plt.figure(figsize=(20,11))
sns.lineplot(x='JobRole',y='MonthlyIncome',data=df)
```

```python
fig,ax=plt.subplots(figsize=(20,10))

sns.lineplot(x='JobRole',y='MonthlyIncome',data=df,ax=ax,hue='Gender',err_style='bars').set(xlabel='JobRole',ylabel='MonthlyIncome',title='JobRole Vs. MonthlyIncome')
```

```python
f,ax=plt.subplots()
new_df=df['Gender'].groupby(df['Attrition']).value_counts(normalize='True').rename('Percentage of group').reset_index()
sns.barplot(x='Attrition',y='Percentage of group',hue='Gender',data=new_df)
vals=ax.get_yticks()
ax.set_yticklabels(['{:,.0%}'.format(x) for x in vals])
```
