```python
""" The subplot() Function
The subplot() function takes three arguments that describes the layout of the figure.

The layout is organized in rows and columns, which are represented by the first and second argument.

The third argument represents the index of the current plot.

plt.subplot(1, 2, 1)
#the figure has 1 row, 2 columns, and this plot is the first plot.

plt.subplot(1, 2, 2)
#the figure has 1 row, 2 columns, and this plot is the second plot.
"""

import pandas as pd
import matplotlib.pyplot as plt
grid = plt.GridSpec(3,3)

fig = plt.figure(figsize=(10,10))
fig.add_subplot(grid[0,:2])
plt.title('G1')
fig.add_subplot(grid[1,:2])
plt.title('G2')
fig.add_subplot(grid[2,:])
plt.title('G3')
fig.add_subplot(grid[:2,2])
plt.title('G4')

```

```python
import pandas as pd
dict1 = {'Name':['100% Bran', '100% Natural Bran', 'All Bran', 'All bran with extra fiber','Almond delight', 'Apple jacks','Basic 4'],'Calories': [70,120,70,50,110,110,130],'Sodium':[130,15,260,140,200,125,210],'Potass':[280,135,320,330,-1,30,100],'Rating':[68.40,33.98,59.42,93.70,34.38,33.17,37.03]}
data=pd.DataFrame(dict1)
data
```

```python
my_grid = plt.GridSpec(3,3)
my_fig = plt.figure(figsize=(10,10))
my_fig.add_subplot(my_grid[0,:2])
plt.plot(data.Potass)
plt.title('Potass')

my_fig.add_subplot(my_grid[1,:2])
plt.plot(data.Sodium)
plt.title('Sodium')




my_fig.add_subplot(my_grid[:2,2])
plt.barh(data.index,data.Rating)
plt.title('Rating')
```
