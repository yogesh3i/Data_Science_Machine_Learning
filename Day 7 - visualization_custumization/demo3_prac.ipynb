```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd 

stats = pd.read_csv('top50.csv')
print(stats.head())

for x,data in stats.groupby('Genre'):
    plt.barh(x,data['Energy'].mean(),color='#FA9696')
    plt.barh(x,data.Danceability.mean(),color='#C80000',left=data['Energy'])
plt.xlabel('Value')
plt.ylabel('Genre')
plt.title("Energy and danceability as per the Genre")
```

```python
# In stats dataset first three columns are catagorical columns we dont need it. 
# Hence we will start the index for subplot from 1. 

import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 

index=1
plt.figure(figsize=(20,18))

for col in stats.columns[3:]:
    plt.subplot(5,2,index)
    plt.hist(stats[col],color='maroon')
    plt.title(col)
    index+=1
plt.show()
 
```

```python
# Use of gridspec: 

import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 

plt.style.use('ggplot')
stat_fig=plt.figure(figsize=(15,12))
stat_grid=plt.GridSpec(4,5)

stat_fig.add_subplot(stat_grid[:2,:3])

for group,data in stats.groupby('Genre'):
    plt.barh(group,data['Energy'].mean(),color='#cd7a9e')
    plt.barh(group,data['Danceability'],color='black',left=data['Energy'])
plt.xlabel('Value')
plt.ylabel('Genre')
plt.title('Danceability per Genre')

stat_fig.add_subplot(stat_grid[:2,3:])
plt.boxplot(stats.Liveness)
plt.title('Liveness')

stat_fig.add_subplot(stat_grid[2:,1:])

for group,data in stats.groupby('Genre'):
    plt.hist(stats.Popularity,color='#f07a9e')
    plt.xlabel('Popularity')
    plt.ylabel('Frequency')
    plt.title('Popularity')
```
