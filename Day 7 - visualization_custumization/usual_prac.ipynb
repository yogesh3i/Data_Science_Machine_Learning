```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a line plot
plt.plot(x, y, label='sin(x)')

# Add a grid to the plot
plt.grid(True, linestyle='--', color='gray', linewidth=0.5)

# Customize grid properties
# linestyle: '--' for dashed line
# color: 'gray' for gray color
# linewidth: 0.5 for thin line

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot with Grid')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a simple bar chart with error bars
categories = ['A', 'B', 'C', 'D']
values = [3, 8, 5, 12]
errors = [1, 0.5, 2, 1.5]

plt.bar(categories, values, yerr=errors)  # specify error bars using yerr parameter
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Bar Chart with Error Bars')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Create a bar chart with labels and annotations
categories = ['A', 'B', 'C', 'D']
values = [3, 8, 5, 12]

plt.bar(categories, values, color='orange')
plt.xlabel('Category', fontsize=14)  # specify font size for x-axis label
plt.ylabel('Value', fontsize=14)  # specify font size for y-axis label
plt.title('Bar Chart with Labels and Annotations', fontsize=16)  # specify font size for title
plt.xticks(rotation=-45, ha='right') 
```

```python
import matplotlib.pyplot as plt
import numpy as np

# create data for the plots
x = np.arange(0, 10, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(x)

with plt.style.context('dark_background'):
    fig=plt.figure(figsize=(10,8))
    gs=plt.GridSpec(2,2)
    fig.add_subplot(gs[0,:1])
    p1=plt.plot(x,y1,color='maroon')
    fig.add_subplot(gs[1,:1])
    p2=plt.plot(x,y2,color='red')
    fig.add_subplot(gs[:1,1])
    p2=plt.plot(x,y3,color='yellow')

    fig.add_subplot(gs[1:,1])
    p2=plt.plot(x,y4,color='violet')
```
