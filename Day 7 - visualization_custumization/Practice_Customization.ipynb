You can customize a line plot created with Matplotlib in various ways. Here are some common customizations you can apply:

# Line Style:
linestyle parameter: You can specify the line style for the plot using the linestyle parameter in the plot() function. For example:


plt.plot(x, y, linestyle='-', label='solid line')  # Solid line
plt.plot(x, y, linestyle='--', label='dashed line')  # Dashed line
plt.plot(x, y, linestyle=':', label='dotted line')  # Dotted line
plt.plot(x, y, linestyle='-.', label='dash-dot line')  # Dash-dot line

# Line Color:
color parameter: You can specify the color of the line using the color parameter in the plot() function. You can use various color formats such as string names (e.g., 'red', 'blue'), hexadecimal RGB values (e.g., '#FF0000'), RGB tuples (e.g., (1, 0, 0)), and more. 

For example:

# Line Width:
linewidth or lw parameter: You can specify the width of the line using the linewidth or lw parameter in the plot() function.

# Adding Labels, Titles, and Legends:
xlabel(), ylabel(): You can set labels for the x-axis and y-axis using xlabel() and ylabel() functions, respectively.
title(): You can set a title for the plot using the title() function.
legend(): You can add a legend to the plot using the legend() function.


# Axis Limits and Ticks:
xlim(), ylim(): You can set the limits for the x-axis and y-axis using xlim() and ylim() functions, respectively.
xticks(), yticks(): You can set custom tick locations on the x-axis and y-axis using xticks() and yticks() functions, respectively.
xticklabels(), yticklabels(): You can set custom tick labels on the x-axis and y-axis using xticklabels() and yticklabels() functions, respectively.
Grid:

```python
import matplotlib.pyplot as plt

# Create a bar chart with labels and annotations
categories = ['A', 'B', 'C', 'D']
values = [3, 8, 5, 12]

plt.bar(categories, values, color='red',alpha=0.5)
plt.xlabel('Category', fontsize=14)  # specify font size for x-axis label
plt.ylabel('Value', fontsize=14)  # specify font size for y-axis label
plt.title('Bar Chart with Labels and Annotations', fontsize=16)  # specify font size for title
plt.xticks(rotation=45, ha='right')  # specify rotation
```

```python
import numpy as np  
import pandas as pd 
import matplotlib.pyplot as plt 
car= pd.read_csv('Car_sales.csv')
print(car.head())
car.shape

plt.style.use('classic')
yt_car = car[car['Vehicle type']=='Car']['Sales in thousands']
yt_passenger = car[car['Vehicle type']=='Passenger']['Sales in thousands']
plt.plot(np.arange(len(yt_car)),yt_car,linestyle='--',color='red')
plt.plot(np.arange(len(yt_passenger)),yt_passenger,linestyle='-.',color='c',linewidth=4.5)
plt.grid()
plt.xlabel('Vehicle type ')
plt.ylabel('Sales in thousand')
plt.title('Sales in thousand for differrnt vehicle type')
plt.legend(['CAR','PASSENGER'],loc='upper right')
plt.xticks(np.arange(0,150,10))
plt.yticks(np.arange(0,600,50))
```

```python
import numpy as np  
import pandas as pd 
import matplotlib.pyplot as plt 
car= pd.read_csv('Car_sales.csv')
car.head()
car.shape

plt.figure(figsize=(8,8))

plt.style.use('default')
for group, data in car.groupby('Name'):
    plt.barh(group,data['4-year resale value'].mean(),color='c')
    plt.barh(group,data['Prices'].mean(),left=data['4-year resale value'].mean(),color='maroon')
plt.title("Mean of 4-year resale value and price for every manufacturer")
plt.xlabel('Price and 4-year resale value')
plt.ylabel('Manufacturer')
plt.legend(['4-year resale value','Price'])
plt.show()
```

```python
# Customization in pie charts:
import numpy as np  
import pandas as pd 
import matplotlib.pyplot as plt 
cmap= plt.get_cmap('Spectral')
cmap
```

```python
clist = [cmap(i) for i in np.linspace(0,1,len(set(car.Name)))]
clist
```

```python
m_dict = dict()
plt.figure(figsize=(10,8))

for manufacturer in set(car.Name):
    m_dict[manufacturer]=car[car.Name==manufacturer].Name.count()
plt.pie(m_dict.values(),labels=m_dict.keys(),colors=clist,rotatelabels=True,autopct='%2f')
plt.show()
```
