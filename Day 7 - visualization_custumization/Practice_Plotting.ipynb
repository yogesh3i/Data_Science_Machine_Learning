```python
# Multiple liner charts: 
""" We can display more than one chart in the same container by using pyplot.figure() function. 
This will help us in comparing the different charts and also control the look and feel of charts """

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt

x = np.array([3,4,5,6,7])
y = x*2
plt.plot(x,y)
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("My_Plot")

# Now to add the new plt i.e. the another plot but within the same output container 
plt.figure()
x = [1,2,7,4,5]
y = [5,6,8,8,8]
plt.plot(x,y,"-.")
plt.show()
```

```python
# Multiple plots in the same axis: 
import numpy as np 
import matplotlib.pyplot as plt 
x = np.array([1,2,3,4,5])
y = x*2
plt.plot(x,y)

x1 = np.array([6,7,8,9,10])
y1 = x*2

plt.plot(x1,y1,"-.")

plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("MY_plot")
plt.show()
```

```python
# Filling the region between the two lines of the data: So that it will be easy for us to understand the margin of the data.py
# This can be done using pyplot.fill between() function.

import numpy as np 
import matplotlib.pyplot as plt

x = np.array([1,2,3,4])
y = x*2

plt.plot(x,y)

x1 = [2,4,6,8]
y1 = [3,5,7,9]

plt.plot(x, y1, "-.")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.title("MY_plot")

plt.fill_between(x,y,y1,color="green", alpha=0.5)
plt.show()
```

```python
# Bar plots:
""""
A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional 
to the values which they represent. The bar plots can be plotted horizontally or vertically. A bar chart describes the comparisons between the 
discrete categories. One of the axis of the plot represents the specific categories being compared, while the other axis represents the measured 
values corresponding to those categories.

The syntax of the bar() function to be used with the axes is as follows:-

plt.bar(x, height, width, bottom, align)   The function creates a bar plot bounded with a rectangle depending on the given parameters. """"



```

```python
import numpy as np 
import matplotlib.pyplot as plt 
data = {"C":20,"C++":30,"JAVA":40,"PYHTON":35}
Courses = list(data.keys())     # Here we have made a dictionary with keys and values so by using data.kes() we are making list of courses and the values assign to it.
Values = list(data.values())  

plt.bar(Courses,Values,color= "maroon",width= 0.4)

plt.xlabel("Course_Offered")
plt.ylabel("No_of_student_enrolled")
plt.title("The_vertical_bar+plot")
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
X = ['A','B','C','D']
Y1 = np.array([12,32,10,25])
Y2 = np.array([23,12,15,16])
Y3 = np.array([15,12,13,24])
Y4 = np.array([14,34,12,26])

plt.bar(X,Y1,color='r')
plt.bar(X,Y2,bottom = Y1, color='b')
plt.bar(X, Y3,bottom=Y1+Y2, color ='y' )
plt.bar(X, Y4, bottom=Y1+Y2+Y3,color ='g')
plt.xlabel('Teams')
plt.ylabel('Score')
plt.legend(['Round1','Round2','Round3','Round4'])
plt.title("Score by Teams in 4 Rounds")
plt.show()
```

```python
# Pie chart: 
""" Pie chart is used to represent a group of a values as part of whole.
Each value is represented as slice of a pie.

A Pie Chart is a circular statistical plot that can display only one series of data. The area of the chart is the total percentage of 
the given data. The area of slices of the pie represents the percentage of the parts of the data. The slices of pie are called wedges.
The area of the wedge is determined by the length of the arc of the wedge. The area of a wedge represents the relative percentage of that 
part with respect to whole data. Pie charts are commonly used in business presentations like sales, operations, survey results, resources, 
etc as they provide a quick summary.


pie() function is used to plot the pie chart.
Pie chart is used to represents the grouped or catagorized data.
using explode() a slice can be separated from the pie chart.
The created list should be equal to the length of the data and 1 represent the value to explode and rest all are 0. 

Pie chaet- Formating: 
The parameter rotateble is used to rotate labels concerning the pie chart.
The parameter autopct is used to show the percentage of each and every slice.

https://www.geeksforgeeks.org/plot-a-pie-chart-in-python-using-matplotlib/


Syntax: matplotlib.pyplot.pie(data, explode=None, labels=None, colors=None, autopct=None, shadow=False)
Parameters: 
data represents the array of data values to be plotted, the fractional area of each slice is represented by data/sum(data). 

If sum(data)<1, then the data values returns the fractional area directly, thus resulting pie will have empty wedge of size 1-sum(data). 

labels is a list of sequence of strings which sets the label of each wedge. 

color attribute is used to provide color to the wedges. 

autopct is a string used to label the wedge with their numerical value. 

shadow is used to create shadow of wedge. 
```

```python
# Scatter plot: 
""""
https://pythongeeks.org/python-scatter-plot/#:~:text=Scatter%20plot%20in%20Python%20is%20one%20type%20of,set%20of%20arrays%20represents%20the%20y%20axis%20data.

Scatter plot in Python is one type of a graph plotted by dots in it. The dots in the plot are the data values. To represent a scatter plot, 
we will use the matplotlib library. To build a scatter plot, we require two sets of data where one set of arrays represents the x axis and the 
other set of arrays represents the y axis data.

- Scatter plot represent the dots as collection of points.
- Order of data in scatter plot is insignificant. 
- Aslo know as collextion plot.


Syntax
matplotlib.pyplot.scatter(xaxis_data, yaxis_data, s = None, c = None, marker = None, cmap = None, vmin = None, vmax = None, alpha = None, linewidths = None, edgecolors = None)

Parameter	Description
xaxis_data	X axis data in an array
yaxis_data	Y axis data in an array
s	        The marker size and it can be scalar or equal to the size of x or y array.
c	        Color for the markers
marker	    Style of the marker
cmap	    Name of the cmap
linewidths	The width of the marker border
edgecolor	Color of the marker border 
Alpha 	    Transparency value which lies between 0 and 1. 0 represents transparent and 1 represents opaque. 



```

```python
# Box plot: To get the range of the perticular data.

# In this we will see the range of hte curb weight.

car['Curb weight'].isna()  # In this its checking for the whole column and displaying it, in such case it will take time to check evry row in the result hence to avoid that we can use the below function .any()

car['Curb weight'].isna().any()

# Hence there are null values to fill that.

car['Curb weight']= car['Curb weight'].fillna(car['Curb weight'].mean())

plt.boxplot(car['Curb weight'])
```

```python
# To see the quartile values 

car['Curb weight'].quantile(0.25)
car['Curb weight'].quantile(0.50)

# To get the IQR (interquartile range)

IQR = car['Curb weight'].quantile(0.75) - car['Curb weight'].quantile(0.25)
high = car['Curb weight'].quantile(0.70) + 1.5*IQR
LOW = car['Curb weight'].quantile(0.25) - 1.5*IQR
print('High = ',high)
print('Low = ',LOW)
print('IQR=' ,IQR)
```
